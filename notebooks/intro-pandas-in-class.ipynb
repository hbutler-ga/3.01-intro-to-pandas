## Handling missing values and duplicates

```python
df.drop_duplicates()
```

```python
print("Before: ", df.shape)

df = df.drop_duplicates()

print("After: ", df.shape)
```

```python
# df.reset_index()
```

Checking for missing values, strategies to deal with them

```python
df.isna().sum()
```

```python
# first strategy is dropping missing values
# df.dropna() will drop any row where there's a missing value in any column

df = df.dropna(subset=['unit_price']) # only drop rows if there's a missing value in a specified col
```

```python
df
```

```python
# df['col'].fillna("value to replace missing values with")

df["region"] = df["region"].fillna("Unknown")
```

```python
df
```

```python
# assuming we Hadn't dropped the row with the missing value in `unit_price`

df['unit_price'] = df['unit_price'].fillna(df['unit_price'].median())
```

```python
df["region"].value_counts()
```

```python
df['region'] = df['region'].str.strip()

df['region'].value_counts()
```

```python
df["status"].value_counts()
```

```python
df["status"] = df["status"].str.strip().str.title()

df["status"].value_counts()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df
```

```python
# create a new column, df['name_of_new_col'] = value

df['order_month'] = df['order_date'].dt.month
```

```python
df['order_day_of_week'] = df['order_date'].dt.day_name()
```

```python
df['revenue'] = df['quantity'] * df['unit_price']

df
```

```python
# to filter our dataset, df[df['col'] == 'some_condition']
df[df['quantity'] >= 5]
```

```python
df[df['region'] == 'Midwest']
```

```python
# filter on multiple conditions
# df[(condition_1) & (condition_2)]

df[(df['region'] == 'Midwest') & (df['quantity'] > 2)]
```

```python
# store our filtered dataframe in a new variable

df_complete = df[df['status'] == 'Complete'].copy() # .copy() creates an independent copy

```
